```python
# Run a query and load result into DataFrame
query = "SELECT * FROM Sales_Analytics.online_retail_ii LIMIT 10;"  # change table name as needed
df = pd.read_sql(query, con=engine)
df
```

```python
df.info()
df.describe()
df.isnull().sum()
df.drop_duplicates(inplace=True)
```

```python
import pandas as pd

# Assume you already loaded the data into df
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.dropna(subset=['Customer ID'], inplace=True)
df['Customer ID'] = df['Customer ID'].astype(int)
df['Total'] = df['Quantity'] * df['Price']

# Rename columns to match SQL
df.rename(columns={
    'Invoice': 'InvoiceNo',
    'Price': 'UnitPrice',
    'Customer ID': 'CustomerID'
}, inplace=True)
```

```python
monthly_sales = df.groupby(df['InvoiceDate'].dt.to_period('M'))['Total'].sum()
monthly_sales.plot(kind='line', title='Monthly Sales')
```

```python
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='bar', title='Top 10 Selling Products')
```

```python
country_sales = df.groupby('Country')['Total'].sum().sort_values(ascending=False).head(10)
country_sales.plot(kind='bar', title='Sales by Country')
```

```python
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (df['InvoiceDate'].max() - x.max()).days,
    'InvoiceNo': 'nunique',
    'Total': 'sum'
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']
print(rfm.head())
```
